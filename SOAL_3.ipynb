```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread('FOTO.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
enhanced = clahe.apply(gray)


_, mask = cv2.threshold(enhanced, 60, 255, cv2.THRESH_BINARY_INV)

brightened = cv2.add(enhanced, 40)
focused = np.where(mask == 255, brightened, enhanced)

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1), plt.imshow(gray, cmap='gray'), plt.title('Grayscale')
plt.subplot(1, 3, 2), plt.imshow(enhanced, cmap='gray'), plt.title('CLAHE Enhanced')
plt.subplot(1, 3, 3), plt.imshow(focused, cmap='gray'), plt.title('Profil Difokuskan')
plt.show()
```
