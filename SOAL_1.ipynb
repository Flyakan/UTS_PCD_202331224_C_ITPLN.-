```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca dan konversi ke RGB
image = cv2.imread('NAMA.jpg')
if image is None:
    raise ValueError("Gambar tidak ditemukan. Pastikan sudah di-upload sebagai 'NAMA.JPG'.")
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Ekstrak channel
r = rgb[:, :, 0]
g = rgb[:, :, 1]
b = rgb[:, :, 2]

# Deteksi masing-masing warna dengan masking
red_mask = np.where((r > g) & (r > b), r, 0)
green_mask = np.where((g > r) & (g > b), g, 0)
blue_mask = np.where((b > r) & (b > g), b, 0)

# Plot 2x2 seperti contoh
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].imshow(rgb)
axs[0, 0].set_title('CITRA KONTRAS')
axs[0, 0].axis('off')

axs[0, 1].imshow(blue_mask, cmap='gray')
axs[0, 1].set_title('BIRU')
axs[0, 1].axis('on')

axs[1, 0].imshow(red_mask, cmap='gray')
axs[1, 0].set_title('MERAH')
axs[1, 0].axis('on')

axs[1, 1].imshow(green_mask, cmap='gray')
axs[1, 1].set_title('HIJAU')
axs[1, 1].axis('on')

plt.tight_layout()
plt.show()
```
